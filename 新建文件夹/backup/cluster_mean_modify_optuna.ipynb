```python
import numpy as np
import pandas as pd
import joblib

from utils import kiwi_operators as kwo
from utils import config
from utils import data_environment as env
from importlib import reload
from glob import glob
import os
import shutil
from utils.feature_optimizers.cluster_mean_modifier import ClusterMeanModifier
from utils import backtester
import optuna

reload(kwo)
reload(config)
```

```python
valid_date_beg = 20160101
valid_date_end = 20210701

path_lst = glob(f"{config.dir_selected_feat}/data/*.pkl")
raw_feat = pd.read_pickle(path_lst[4]).dropna(how="all").fillna(method="ffill")
raw_feat = kwo.get_partial_dataframe_by_date(raw_feat, valid_date_beg, valid_date_end)


def objective(trial):
    num_rows = trial.suggest_int("num_rows", 5, 90)
    n_clusters = trial.suggest_int("n_clusters", 2, 10)
    mean_ratio = trial.suggest_float("mean_ratio", 0.1, 0.9)
    modifier = ClusterMeanModifier(
        num_train_rows=num_rows, n_clusters=n_clusters, mean_ratio=mean_ratio
    )
    mod_feat = modifier.transform(raw_feat)
    raw_sharpe = backtester.calc_sharpe(raw_feat)
    mod_sharpe = backtester.calc_sharpe(mod_feat)
    return mod_sharpe - raw_sharpe


SEED = 666
study = optuna.create_study(
    direction="maximize",
    sampler=optuna.samplers.TPESampler(seed=SEED),
)

study.optimize(objective, n_trials=1000, timeout=6000)

print(study.best_value, flush=True)
print(study.best_params, flush=True)
```
